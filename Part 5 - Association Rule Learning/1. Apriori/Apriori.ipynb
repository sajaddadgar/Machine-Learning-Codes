# Apriori

**1. Data Preprocessing**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataframe = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
# each row: one user
dataframe.head(5)
```

**1.2. Modify Dataframe to Use as Inputs of Apriory Algorithm** (2 for loops: users and their basket)

```python
transactions = []
```

```python
for i in range(0, int(dataframe.shape[0])):
    transactions.append([str(dataframe.values[i, j]) for j in range(dataframe.shape[1])])
```

**2. Apriori Algorithm**

```python
from apyori import apriori
```

**Minimum Support**: products that purchase 3 times in week = (3 x 7) / (7500) = 0.0028 < 0.003<br>
**For Calculate Minimum Support:** Change Values to Get Good Results

```python
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

```python
# Sort by Relevance (combination of support and confidence and lift)
rules = list(rules)
```

```python
print('Association Rule: {}'.format(rules[0][0]))
print('support: {}'.format(rules[0][1]))
print('other: {}'.format(rules[0][2]))
```
